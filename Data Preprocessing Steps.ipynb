# Preporcessig Steps

## 1. Lowercasing

```python
new_df['title'] = new_df['title'].apply(lambda x:x.lower())
new_df['overview'] = new_df['overview'].apply(lambda x:x.lower())
```

```python
new_df.head()
```

## 2. Remove HTML Tags

```python
import re
def remove_html_tags(text):
    pattern = re.compile('<,*?>')
    return pattern.sub(r'', text)
```

```python
new_df['overview'] = new_df['overview'].apply(remove_html_tags)
```

## 3. Remove URLs

```python
def remove_url(text):
    pattern = re.compile('https?://\S+|www\.S+')
    return pattern.sub(r'', text)
```

```python
new_df['overview'] = new_df['overview'].apply(remove_url)
```

## 4. Removing Punctuation 

```python
import string
exclude = string.punctuation
```

```python
def remove_punct(text):
    return text.translate(str.maketrans('','',exclude))
```

```python
new_df['title'] = new_df['title'].apply(remove_punct)
new_df['overview'] = new_df['overview'].apply(remove_punct)
```

## 5. Remove Stopwords

```python
from nltk.corpus import stopwords
stopwords = stopwords.words('english')
```

```python
def remove_stopword(text):
    tokens = text.split()
    new_text = []
    for token in tokens:
        if token not in stopwords:
            new_text.append(token)
    return ' '.join(new_text)
```

```python
new_df['title'] = new_df['title'].apply(remove_stopword)
new_df['overview'] = new_df['overview'].apply(remove_stopword)
```

## 6. Stemming/Lemmitization

Stemming - Stemming means mapping a group of words to the same stem by removing prefixes or suffixes without giving any value to the “grammatical meaning” of the stem formed after the process. 

Stemmer - It is algorithum to do stemming
1. Porter Stemmer - specific for english language
2. Snowball Stemmer - use for multiple language
3. Lancaster Stemmer

Lemmitization - Lemmatization also does the same thing as stemming and try to bring a word to its base form, but unlike stemming it do keep in account the actual meaning of the base word.

Stemming is fast than lemmitizer but in stemming the stem word may or may not have meaning and in case in lemmitizer word have always meaning.

Here I have use Lemmitizer, we case stemming as well.

```python
from nltk.stem import WordNetLemmatizer
```

```python
lemma = WordNetLemmatizer()
```

```python
def lemmitizer(text):
    tokens = text.split()
    new_text = []
    for token in tokens:
        new_text.append(lemma.lemmatize(token,'v'))
    return ' '.join(new_text)
```

```python
new_df['overview'] = new_df['overview'].apply(lemmitizer)
```
